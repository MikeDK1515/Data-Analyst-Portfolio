```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pokemon = pd.read_csv("Pokemon_stats.csv")
```

```python
pokemon.describe()
```

```python
pokemon.dtypes
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12,8))
sns.histplot(pokemon['Attack'])
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12,8))
sns.histplot(pokemon['Defense'])
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12,9))
sns.histplot(pokemon['Speed'])
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
#Replace 'Type 1' with the names of the other columns for each graph
ax = sns.countplot(x='Type 1', data=pokemon, order=pokemon['Type 1'].value_counts().index)
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))

ax = sns.countplot(x='Type 2', data=pokemon, order=pokemon['Type 2'].value_counts().index)
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))

ax = sns.countplot(x='Generation', data=pokemon, order=pokemon['Generation'].value_counts().index)
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(pokemon[['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']].corr(), vmin=-1, vmax=1, annot=True, annot_kws={"fontsize":18}, cmap="PiYG")
```
