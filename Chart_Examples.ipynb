```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
wines = pd.read_csv("wines_cleaned.csv")
wines.describe()
```

```python
wines.dtypes
```

```python
wines['type'].value_counts()
```

```python
wines.head()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,8))
sns.histplot(wines['pH'])
```

```python
sns.histplot(wines['pH'], bins=12)
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(12,10))
ax = sns.countplot(x='type', data=wines)
```

```python
wines['type'].value_counts()
```

```python
ax = sns.countplot(x='type', data=wines, order = wines['type'].value_counts().index)
```

```python
plt.figure(figsize=(10,10))

sns.heatmap(wines.corr(), vmin=-1, vmax=1, annot=True, annot_kws={"fontsize":14}, cmap="PiYG")
```

```python
plt.plot(wines.alcohol, wines.quality)
plt.title('Wines alcohol content by quality')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.show

```

```python
data = [["Monday", 10],["Tuesday", 3],["Wednesday", 5],["Thursday", 12],["Friday", 6],["Saturday",7],["Sunday", 9]]
df = pd.DataFrame(data, columns = ["Day", "Admissions"])
```

```python
plt.plot(df.Day, df.Admissions)
plt.title('Theater admissions by day of the week')
plt.xlabel('Day')
plt.ylabel('Admissions')
plt.show
```
